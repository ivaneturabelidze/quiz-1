```python
# The pie diagram shows which countries developers surveyed from. Top 3 countries are exploded from the diagram
threshold = 1000
y = df['Country'].value_counts()
small_countries = y[y < threshold]
rest_count = small_countries.sum()
y = y[y >= threshold]
y['Rest'] = rest_count
x = y.index
plt.figure(figsize=(15,15))
explode = [0] * len(x)
explode[0:3] = [0.1] * 3
plt.pie(y, labels=x, explode=explode, textprops={'fontsize': 12})
plt.title('Survey participants by countries', fontsize=21)
plt.show()
```

```python
# The bar diagram shows salaries by countries in ascending order
y = df.groupby('Country')['ConvertedCompYearly'].median().sort_values()
x = y.index.tolist()
positions = range(len(x))
plt.title('Yearly income median by countries')
plt.figure(figsize=(500,80))
plt.xticks(positions, x)
plt.bar(positions,y)
```
